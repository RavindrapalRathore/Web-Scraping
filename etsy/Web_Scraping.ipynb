```python
# URL of the Etsy webpage
url = "https://www.etsy.com/in-en/featured/staff-picks-in?ref=hp_g-1"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Get the HTML content of the page
    html_content = response.text

    # Create a BeautifulSoup object to parse the HTML
    soup = BeautifulSoup(html_content, 'html.parser')

    # Extract product information
    product_info = []

    # Find product cards (you may need to adjust the selector based on the website structure)
    product_cards = soup.find_all('div', class_='v2-listing-card')

    for card in product_cards:
        # Extract relevant information from each product card
        product_title_tag = card.find('h3', class_='wt-text-caption v2-listing-card__title wt-text-truncate')
        product_price_tag = card.find('span', class_='currency-value')
        product_originalp_tag = card.find('span', class_='wt-text-strikethrough')
        product_url_tag = card.find('a', class_='listing-link wt-display-inline-block')
        product_delivery_tag = card.find('span', class_='wt-badge wt-badge--small wt-badge--statusValue')
        product_stock_tag = card.find('span', class_='currency-value')
        product_seller_tag = card.find('span', class_='wt-screen-reader-only')

        # Check if the elements exist before attempting to access their text
        if product_title_tag:
            product_title = product_title_tag.text.strip()
        else:
            product_title = 'N/A'

        if product_price_tag:
            product_price = product_price_tag.text.strip()
        else:
            product_price = '0'

        if product_url_tag:
            product_url = product_url_tag.get('href')
        else:
            product_url = 'N/A'

        if product_delivery_tag:
            product_delivery = product_delivery_tag.text.strip()
        else:
            product_delivery = 'Delivery Charged'


        if product_stock_tag:
          product_stock = product_stock_tag.text.strip()
        else:
            product_stock = 'Out Of Stock'


        if product_seller_tag:
          product_seller = product_seller_tag.text.strip()
        else:
            product_seller = 'N/A'


        if product_originalp_tag:
            product_originalp = product_originalp_tag.text.strip()
        elif product_price_tag:
            product_originalp = product_price_tag.text.strip()

            # Modify the product_stock assignment in your loop
        if product_stock.strip().lower() == 'out of stock':
          product_stock = 'Out-Of_Stock'
        else:
          product_stock = 'In-Stock'

        # Modify the product_originalp assignment in your loop
        if product_originalp:
          product_originalp = product_originalp.replace('', '').replace(',', '').strip()
        else:
          product_originalp = 'N/A'


        # Append product data to the list
        product_info.append([product_title, product_price, product_originalp, product_delivery, product_stock, product_url, product_seller])
    # Save the data to a CSV file
    csv_filename = 'etsy_products.csv'

    with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['Product Title', 'Price', 'Original Price', 'Delivery', 'Stock', 'Product URL', 'Seller Name']
        writer = csv.writer(csvfile)

        # Write header row
        writer.writerow(fieldnames)

        # Write product data
        writer.writerows(product_info)

    print(f'Data saved to {csv_filename}')
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```
